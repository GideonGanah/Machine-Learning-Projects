```python
# Here we import all the nescesary libraries
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
#here we upload the dataset file into the program
from google.colab import files
upload = files.upload()
```

```python
# here we call the uploaded data and label it by adding  column names
fileName = "Energy Meter.csv"
names = ['Voltage', 'Current','Power','class']

# here we reead the data
dataset = read_csv(fileName, names=names)
```

```python
# here we assign the dataset to a variable (the data set is seen as a dataframe)
array = dataset.values
#here we set independent variables( x values, they are the inputs)
x= array[:,0:3]
# here we set the dependable varable( y value,, this is the output)
y= array[:,3]
# here we split the dataset into Train and validation (train data and test data).
# the test size in the perenthesis determins the percentage of the data to be used as test data
# since the test_size is 20%, the x_train and y_train is going to hold 80% of the data for training
# x_validation and y_validation will also hold 20% of the data for testing purposes
x_train, x_validation, y_train, y_validation = train_test_split(x,y, test_size = 0.20, random_state=1)
```

```python
# here we go for our model. we call the model that we will use for the classification(Suport Vector Machine)
model = SVC(gamma ='auto') #gamma is a parameter for non-linear hyperplanes
```

```python
# here we insert our data into the model for training
model.fit (x_train, y_train)
```

```python
# here we print the results of model training. The value we get determins the accuracy of the model
results = model.score(x_validation, y_validation)
print(results)
```

```python
# here we plug in data to test the model
value = [[212.7315,0.84753,180.2963282]]
predictions = model.predict(value)
print(predictions[0])
```
