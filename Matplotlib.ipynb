**Double Plot-line**

```python
from matplotlib import pyplot as plt
x = [1,2,3,4,5,6,7,8,9]
y = [10,2,3,6,4,9,7,4,2]
x2 = [1,2,3,4,5,6,7,8,9]
y2 = [2,4,7,9,5,3,8,10,2]



plt.plot(x,y,'r', label='Profit') # this will plot the graph with the line colour red, the label part will create a legend
plt.plot(x2,y2,'g',label='Tax')# this will plot the graph with the line colour Green
plt.plot(x,y,'ro')  # this will plot the graph with dot at eachh coordinate
plt.plot(x2,y2,'go')  # this will plot the graph with  dot at eachh coordinate

plt.legend()
plt.title("Graph")
plt.xlabel("Order")
plt.ylabel("Count")
plt.show()
```

**Pie chart**

```python
from matplotlib import pyplot as plt
x = [80,20,40,50,90,5]
y = ["Math", "physics", "Programming","Communiation","History","Sports"]
c = ['r','g','c','b','m', 'y','w','k']

plt.pie(x,labels=y,colors=c,autopct='%1.1f%%' )

plt.title("Graph")
plt.show()
```

**MULTIPLE GRAPHS IN SAME PROGRAM**

```python
from matplotlib import pyplot as plt
x = [1,2,3,4,5,6,7,8,9]
y = [10,2,3,6,4,9,7,4,2]
x2 = [1,2,3,4,5,6,7,8,9]
y2 = [2,4,7,9,5,3,8,10,2]

plt.subplot(2,1,1,)
plt.title("Graph 1")
plt.plot(x,y,'r')

plt.subplot(2,1,2)
plt.title("Graph 2")
plt.plot(x2,y,'g')
plt.show()
```
