```python
import numpy as np
from sklearn.datasets import load_digits # here we import sklearn dataset.
```

```python
# here we load our dataset
""" Classes               10
    Samples per class     180
    samples total         1797
    Dimensionality        64 (8X8)
    Features              Integers 0-16"""

dataset = load_digits()
```

```python
# here we summarize dataset
print(dataset.data)
print(dataset.target)


print(dataset.data.shape)
print(dataset.images.shape)

dataimageLength = len(dataset.images)
print(dataimageLength)
```

```python
# here we visualize the data
import matplotlib.pyplot as plt  #we import the nnneeded library
n=10   #  number of samples out of sample total 1797

plt.gray()    # we plot a gray image of the data
plt.matshow(dataset.images[n])  # we plot dataset
plt.show()   # we display the chart

dataset.images[n]
```

```python
# here we segeregate our data
x = dataset.images.reshape((dataimageLength, -1))
x
```

```python
#  to get our y values
y = dataset.target
y
```

```python
# here we split our dataset into train and test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state = 0)
print(x_train.shape) # this will help us see the amount of data going into train
print(x_test.shape) # this will help us see the amount of data going into test
```

```python
# training
from sklearn import svm
model = svm.SVC(kernel='linear')
model.fit(x_train,y_train)
```

```python
# pridicting what value is from test Data
n=-1
result = model.predict(dataset.images[n].reshape((1,-1)))

plt.imshow(dataset.images[n], cmap=plt.cm.gray_r, interpolation = 'nearest')
print(result)
print("\n")
plt.axis('off')
```

```python
# DOING prediction with our test data
y_pred = model.predict(x_test)

print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
# HERE we we validate our model
from sklearn.metrics import accuracy_score

print("The accuracy score of the Model: ",accuracy_score(y_test,y_pred)*100)
```

```python
# PLAYING with difffrent methods and parameters

from sklearn import svm
# here we call models
model1 = svm.SVC(kernel ='linear')
model2 = svm.SVC(kernel ='rbf')
model3 = svm.SVC(gamma=0.001)

# here we train models
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)

# here we test models
y_predM1 = model1.predict(x_test)
y_predM2 = model2.predict(x_test)
y_predM3 = model3.predict(x_test)


print("The accuracy score of the Model 1:{0}% ".format(accuracy_score(y_test,y_predM1)*100))
print("The accuracy score of the Model 2:{0}% ".format(accuracy_score(y_test,y_predM2)*100))
print("The accuracy score of the Model 3:{0}% ".format(accuracy_score(y_test,y_predM3)*100))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model4 = KNeighborsClassifier(n_neighbors=5, metric='mekowski', p=3)
model4.fit(x_train,y_train)
result = model4.predict(x_test)

print("Accuracy : {0}%".format(accuracy_score(y_test,result)*100))
```
