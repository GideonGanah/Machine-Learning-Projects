**Generating** **Array**

```python
c= np.arange(10) #range of values in array
d= np.linspace(1,5,10) #set of values by interval [start,end,no.ofvaluesEqualInterval]
print(c)
print(d)
```

Array properties

```python
a = np.array([7,9,9,2])
b = np.array([(7,9,9,2),(16,11,9,6)])
print(a.ndim) #shows Dimension of array
print(a.itemsize) #shows the byte size of each element
print(a.dtype)   #show data type of array
print(a.size*a.itemsize)   #shows memory occuppoed i.e space occuppiedd by 1 element * length of numpy array
print(b.size) # shows number of elements in array
print(b.shape) #shows shape of array(rows and columns)

```

**Array Reshaping**

```python
b = np.array([(7,9,9,2),(16,11,9,6)])
print(b)
b=b.reshape(4,2)
print(b)
```

**Array Slicing**

```python
b = np.array([(7,9,9,2),(16,11,9,6)])
print(b[0,2]) # slicing the element from certain row and certain column b[rowIndex,ElementIndex]
print(b[0:,3]) #slcing element from all rows and certain collumn
```

**Array Value operation**

```python
b = np.array([(7,9,9,2,16,11,9,6)])
print(b.min())  #minimum value in Array
print(b.max())  #maximum value in Array
print(b.sum()) #Sum of Array
```

**Array Arithmetic Operation**

```python
a = np.array([7,9,9,2])
b = np.array([1,9,9,6])
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
 a = np.array([7,9,9,2])
 b = np.array([(7,9,9,2,16,11,9,6)])
 print(b.sum(axis=0))   #sums each elememtn of 1st dimention array with the same index with another dimention
 print(b.sum(axis=1))   #sums of all elements in 1st dimention & sum of all elements in 2nd dimention
 print(np.sqrt(a)) #finds square root of each element
 print(np.std(a))  #finds standard diviation of the array
 print(np.exp(a))  #finds exponential of the array
 print(np.log(a))  # natural log of array -In
 print(np.log10(a)) #log base 10
```

```python
 a = np.array([(7,9,9,2),(16,11,9,6)])
 b = np.array([(1,2,3,4),(5,6,7,8)])
 print(np.vstack((a,b)))  # vertical stacking of arrays
 print(np.hstack((a,b)))  #Horizontal stack
 print(a.ravel()) #merges dimenntions in array into a single dimention
```
